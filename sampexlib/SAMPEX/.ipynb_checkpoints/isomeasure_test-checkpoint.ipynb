```python
%matplotlib inline
import math as m
import numpy as np
import matplotlib.pyplot as plt
from SAMPEX_functions import read_counts as read
from SAMPEX_functions import mb_finder, iso_calculator
import pandas as pd
```

```python
# reading in electron counts file: day ### of year ####

year = '1993'
start_day = '308'

file = 'hhrr' + year + start_day + '.txt'
t, r1, r2, r3, r4 = read(file)
```

```python
print(len(r1))
print(len(r4))
```

```python
# calculation of the isotropy indices of electron counts (integer method)

iso_indices = []        # for data analysis purposes
iso_indices_plot = []   # for plotting purposes

for i in np.arange(len(r1)):
    if r4[i] >= 20 and r1[i] >= 20:
        if  r4[i] != 0 and r4[i] > r1[i]:
            iso_index = r1[i]/r4[i]
            iso_indices.append(iso_index)
            iso_indices_plot.append(iso_index)
        elif  r1[i] != 0 and r1[i] >= r4[i]:
            iso_index = r4[i]/r1[i]
            iso_indices.append(iso_index)
            iso_indices_plot.append(iso_index)
    else:
        # when counts are too low, they do not contribute to the isotropy day-average isotropy index
        iso_index = 0
        iso_indices_plot.append(iso_index)
        
iso_indices = np.array(iso_indices)             # only the iso indices true values
iso_indices_plot = np.array(iso_indices_plot)   # returns 0 for indices that do not pass the 20 count test (for plotting)
```

```python
plt.figure(figsize = (30,15))
plt.title('SSD1/SSD4 Ratios vs Time (Integer Calc)', fontsize = 30)
plt.scatter(t, iso_indices_plot)
plt.xticks(fontsize = 25)
plt.yticks(fontsize = 25)
plt.xlabel('Time (seconds)', fontsize = 25)
plt.ylabel('Isotropy (1-scale)', fontsize = 25)
plt.ylim(0.45, 1.05)
plt.grid(True)
plt.show()
```
