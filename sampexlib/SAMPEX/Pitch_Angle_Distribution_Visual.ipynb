```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set ansio equation
def aniso_plot(x):
    #y = np.sign(x) * (np.abs(x)) ** (1 / 3)
    sig = len(x)/1.9
    mu = len(x)
    y = ( sig*np.sqrt(2*np.pi) * np.exp( ((x-mu)/sig)**2) / 2) ** -1
    y_scaled = (y * 4e5/np.max(y))
    y_scaled =  y_scaled - np.min(y_scaled)
    return y_scaled

# set iso equation
def iso_plot(x, h):
    y = np.array([])
    for i in x:
        y = np.append(y, h)
    return y
```

```python
x = np.arange(91)

y_iso = iso_plot(x, 1.2e5)
y_aniso = aniso_plot(x)

x_fill = np.arange(0, 8, 0.1)
```

```python
plt.figure(figsize = (10,8))
plt.title(f'Isotropic & Anisotropic Electron Pitch Angle Distributions', fontsize = 15)
plt.plot(x, y_iso, color='navy')
plt.plot(x, y_aniso, color='darkorange')
plt.fill_between(x_fill, 0, 4e5, facecolor='maroon', alpha=0.25)
plt.vlines(np.max(x_fill), 0, 4e5, colors='maroon', linestyle='dashed', alpha=0.5)
plt.xlim(0, 90)
plt.ylim(0, 4e5)
plt.xlabel('Equatorial Pitch Angle [degrees]', fontsize=14)
plt.ylabel('Electron Counts', fontsize=14)
plt.legend(['Isotropic', 'Anisotropic', 'Loss Cone'], loc='lower right', fontsize = 15)
plt.grid(alpha=0.5, zorder=1)
```
